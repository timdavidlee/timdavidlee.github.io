# Sourcing Data

### All cities in Craigslist

https://www.craigslist.org/about/sites

This makes things a lot easier, beause it lists all the city prefixes. We will use this to pull all the city prefixes

```python
import requests
from scrapy.selector import Selector

response = requests.get('https://www.craigslist.org/about/sites')
cllist = Selector(text =response.text).xpath("//div[@class='colmask'][1]/div/ul/li/a/@href").extract()
cllist
```

### Types of electronics that will be researched

```python
searchTerms = [
    'search/sss?sort=rel&query=macbook'
    ,'search/sss?sort=rel&query=iphone'
    ,'search/sss?sort=rel&query=samsung+galaxy'
]
```

```python
bb = ['http:'+str(x)+y for x in cllist for y in searchTerms]
```

### Create starting page links to crawl

```python
for b in bb:
    print '"'+b+'",'
```

# Process Scraped Data

### Load Libraries

```python
import json
import cPickle
from altair import Chart
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

### Load Scraped Datasets

```python
filelist = [
    "items2.json"
    ,'bot1_results.json'
    ,'bot2_results.json'
    ,'bot3_results.json'
    ]
```

### Consolidating all files into single dataframe

```python
df = pd.concat( [pd.read_json(x) for x in filelist])
```

```python
df.info()
```

```python
df.head(3)
```

### Data cleaning

```python
clean_df = df.copy()
```

```python
print clean_df.shape
clean_df['price'].isnull().sum()
```

```python
#dropping blank titles and blank prices
clean_df.drop(clean_df[clean_df['price'].isnull()].index,inplace=True)
clean_df.drop(clean_df[clean_df['title'].isnull()].index,inplace=True)
```

```python
clean_df['price'].isnull().sum()
```

```python
clean_df.shape
```

```python
# check the price section
b = clean_df['price'].values
for i,x in enumerate(b):
    if type(x) != list:
        print i, type(x), x
```

```python
clean_df['clean_price'] = clean_df['price'].map(lambda x : float(x[0].replace('$','')))
clean_df['clean_lat'] = clean_df['latitude'].map(lambda x : x[0] if type(x) == list else x)
clean_df['clean_long'] = clean_df['longitude'].map(lambda x : x[0] if type(x) == list else x)

clean_df['macbook'] = clean_df.apply(lambda x : 1 if 'macbook' in x['title'][0].lower() else 1 if 'macbook' in x['body'].lower() else 0 ,axis=1)
clean_df['iphone'] = clean_df.apply(lambda x : 1 if 'iphone' in x['title'][0].lower() else 1 if 'iphone' in x['body'].lower() else 0 ,axis=1)
clean_df['galaxy'] = clean_df.apply(lambda x : 1 if 'galaxy' in x['title'][0].lower() else 1 if 'galaxy' in x['body'].lower() else 0 ,axis=1)

clean_df['category'] = clean_df.apply(lambda x : 'macbook' if x['macbook'] == 1 else 'iphone' if x['iphone'] == 1 else 'galaxy' if x['galaxy']==1 else 'none' ,axis=1)
```

```python
clean_df['category'].value_counts()
```

```python
clean_df.head(3)
```

```python
clean_df[clean_df['category']=='none'].head()
clean_df.drop(clean_df[clean_df['category']=='none'].index,inplace=True)

#dropping extreme outliers
clean_df.drop(clean_df[clean_df['clean_price']>3000].index,inplace=True)
clean_df.drop(clean_df[clean_df['clean_price']<10].index,inplace=True)

iphone_df = clean_df[clean_df['category']=='iphone'].copy()
galaxy_df = clean_df[clean_df['category']=='galaxy'].copy()
macbook_df = clean_df[clean_df['category']=='macbook'].copy()

#trimming deviations

# ============================================================================
# iphone 
# ============================================================================
mean = iphone_df['clean_price'].mean()
std = iphone_df['clean_price'].std()
ubound = mean+2*std
lbound = mean-2*std

# only looking at prices within 2 deviations
iphone_df.drop(iphone_df[iphone_df['clean_price']>ubound].index,inplace=True)
iphone_df.drop(iphone_df[iphone_df['clean_price']<lbound].index,inplace=True)


# ============================================================================
# galaxy 
# ============================================================================
mean = galaxy_df['clean_price'].mean()
std = galaxy_df['clean_price'].std()
ubound = mean+std
lbound = mean-std

# only looking at prices within 2 deviations
galaxy_df.drop(galaxy_df[galaxy_df['clean_price']>ubound].index,inplace=True)
galaxy_df.drop(galaxy_df[galaxy_df['clean_price']<lbound].index,inplace=True)

# ============================================================================
# macbook 
# ============================================================================
mean = macbook_df['clean_price'].mean()
std = macbook_df['clean_price'].std()
ubound = mean+std
lbound = mean-std

# only looking at prices within 2 deviations
macbook_df.drop(macbook_df[macbook_df['clean_price']>ubound].index,inplace=True)
macbook_df.drop(macbook_df[macbook_df['clean_price']<lbound].index,inplace=True)

```

### Additional Detail

```python
iphone_df['model'] = iphone_df['title'].map(lambda x : re.compile('iphone\s?\d[s]?\s?').findall(x[0].lower() ))
iphone_df['model'] = iphone_df['model'].map(lambda x : x[0] if len(x)==1 else x)
iphone_df['model'] = iphone_df['model'].map(lambda x : '' if type(x)==list else str(x.encode('utf-8')).strip().replace(' ',''))
iphone_df['len_model'] = iphone_df['model'].map(lambda x : type(x))
iphone_df.head()
```
